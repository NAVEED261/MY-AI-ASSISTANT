```python
# Import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
# Creating a simple dataset for house prices
data = {
    'Size (sq ft)': [500, 1000, 1500, 2000, 2500, 3000, 3500],
    'Price (in $1000)': [100, 200, 300, 400, 500, 600, 700]
}

# Converting data to DataFrame
df = pd.DataFrame(data)
print(df)
```

```python
# Visualizing the data
plt.scatter(df['Size (sq ft)'], df['Price (in $1000)'], color='blue')
plt.title('House Prices vs Size')
plt.xlabel('Size (sq ft)')
plt.ylabel('Price (in $1000)')
plt.show()
```

```python
# Features (X) aur Target (y) ko divide karo
X = df[['Size (sq ft)']]  # Features (Input)
y = df['Price (in $1000)']  # Target (Output)

# Data ko train aur test sets mein split karo (80% training, 20% testing)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# Linear Regression model banake usko train karo
model = LinearRegression()
model.fit(X_train, y_train)

# Slope (coefficient) aur intercept print karo
print("Model Coefficient (Slope):", model.coef_)
print("Model Intercept:", model.intercept_)
```

```python
# Predictions banaye model se
y_pred = model.predict(X_test)

# Actual prices aur predicted prices ka comparison
comparison = pd.DataFrame({'Actual Price': y_test, 'Predicted Price': y_pred})
print(comparison)
```

```python
# Regression line plot karo (actual vs predicted)
plt.scatter(X_test, y_test, color='blue')  # Actual data points
plt.plot(X_test, y_pred, color='red')  # Regression line (prediction)
plt.title('Actual vs Predicted Prices')
plt.xlabel('Size (sq ft)')
plt.ylabel('Price (in $1000)')
plt.show()
```

```python
from sklearn.metrics import mean_squared_error

# MSE calculate karo
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)
```

```python
# Importing necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Creating a simple dataset for house prices
data = {
    'Size (sq ft)': [500, 1000, 1500, 2000, 2500, 3000, 3500],
    'Price (in $1000)': [100, 200, 300, 400, 500, 600, 700]
}

# Converting data to DataFrame
df = pd.DataFrame(data)

# Features (X) and Target (y)
X = df[['Size (sq ft)']]  # Features (Input)
y = df['Price (in $1000)']  # Target (Output)

# Splitting data into training and testing sets (80% training, 20% testing)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Creating and training the Linear Regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Making predictions using the model
y_pred = model.predict(X_test)

# Actual vs Predicted Prices comparison
comparison = pd.DataFrame({'Actual Price': y_test, 'Predicted Price': y_pred})

# Calculating MSE
mse = mean_squared_error(y_test, y_pred)

# Plotting the regression line
plt.scatter(X_test, y_test, color='blue')  # Actual data points
plt.plot(X_test, y_pred, color='red')  # Regression line (prediction)
plt.title('Actual vs Predicted Prices')
plt.xlabel('Size (sq ft)')
plt.ylabel('Price (in $1000)')
plt.show()

comparison, mse
```
