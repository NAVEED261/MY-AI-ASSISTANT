```python
# Import necessary libraries
import pandas as pd
from sklearn.ensemble import IsolationForest
from sklearn.preprocessing import LabelEncoder
```

```python
# Sample data URL for network traffic
url = "https://raw.githubusercontent.com/defcom17/NSL_KDD/master/KDDTrain+.txt"
columns = ['duration', 'protocol_type', 'service', 'flag', 'src_bytes', 'dst_bytes', 'wrong_fragment', 'urgent', 'label']

# Load the dataset into pandas dataframe
data = pd.read_csv(url, names=columns)

# Display the first few rows
data.head()
```

```python
# Convert categorical columns to numeric using LabelEncoder
label_encoder = LabelEncoder()

# Encoding protocol_type, service, and flag
data['protocol_type'] = label_encoder.fit_transform(data['protocol_type'])
data['service'] = label_encoder.fit_transform(data['service'])
data['flag'] = label_encoder.fit_transform(data['flag'])

# Extract the features (X) and labels (y)
X = data[['duration', 'protocol_type', 'service', 'flag', 'src_bytes', 'dst_bytes']]
y = data['label']  # This will have 'normal' and 'attack'
```

```python
# Train the Isolation Forest model
model = IsolationForest(contamination=0.1, random_state=42)  # Assume 10% of the data is suspicious
model.fit(X)

# Predict whether each record is an anomaly (-1 means suspicious)
predictions = model.predict(X)

# Map predictions to 'Normal' and 'Suspicious'
data['anomaly'] = ['Normal' if x == 1 else 'Suspicious' for x in predictions]

# Display the results
data[['duration', 'protocol_type', 'service', 'flag', 'src_bytes', 'dst_bytes', 'anomaly']].head()
```

```python
# Check how many normal and suspicious activities were detected
print(data['anomaly'].value_counts())
```

```python
# Assume src_ip and dst_ip columns are present in the data (or add them manually)
data['src_ip'] = ['192.168.1.' + str(i % 100) for i in range(len(data))]  # Example source IPs
data['dst_ip'] = ['10.0.0.' + str(i % 50) for i in range(len(data))]     # Example destination IPs

# Filter the data to show only suspicious activities
suspicious_traffic = data[data['anomaly'] == 'Suspicious']

# Show details of the suspicious traffic
report_columns = ['src_ip', 'dst_ip', 'protocol_type', 'service', 'flag', 'src_bytes', 'dst_bytes']
suspicious_report = suspicious_traffic[report_columns]

# Display the report
print("Detailed Suspicious Traffic Report:")
suspicious_report.head()
```

```python
import requests

def send_alert(ip_address):
    # Simulate sending an alert to a SIEM tool or webhook (e.g., Splunk)
    alert_message = f"Suspicious activity detected from IP: {ip_address}"
    print(f"Sending alert: {alert_message}")

    # Simulated webhook URL (replace with actual Splunk or Elastic webhook)
    webhook_url = "https://example.com/webhook"

    # Sending a POST request (simulate alert to Splunk or SIEM)
    requests.post(webhook_url, json={"alert": alert_message})

# Send alert for each suspicious activity detected
for ip in suspicious_report['src_ip']:
    send_alert(ip)
```

```python
import os

# Define the function to block the IP
def block_ip(ip_address):
    # Command to block the IP address using iptables
    command = f'sudo iptables -A INPUT -s {ip_address} -j DROP'

    # Execute the command in the system
    os.system(command)

    # Print confirmation
    print(f"IP Blocked: {ip_address}")

# Example IP to block
suspicious_ip = "192.168.1.80"

# Call the function with the suspicious IP address
block_ip(suspicious_ip)
```
